```python
# Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
# # Distribution and Binomial Distribution?

"""Bernoulli distribution is a discrete probability distribution that models the probability of a single binary event
  The Bernoulli distribution has a single parameter, p, which represents the probability of success. The probability of failure is 1-p.
The probability mass function (PMF) of the Bernoulli distribution is given by:

P(x) = p^x(1-p)^(1-x)

where x can take on values 0 or 1 (0 for failure, 1 for success).

The binomial distribution, on the other hand, is a discrete probability distribution that models the probability of obtaining
a certain number of successes in a fixed number of independent Bernoulli trials. The binomial distribution has two parameters:
   the number of trials (n) and the probability of success (p). The probability mass function (PMF) of the binomial distribution 
   is given by:

P(x) = nCx * p^x * (1-p)^(n-x)

where x can take on values from 0 to n, and nCx  is the binomial coefficient, given by:

nCx = n! / (x! * (n-x)!)

The difference between the Bernoulli distribution and the binomial distribution is that the Bernoulli distribution models the
 probability of a single binary event , while the binomial distribution models the probability of obtaining a certain number of
successes in a fixed number of independent Bernoulli trials. The Bernoulli distribution is
a special case of the binomial distribution, where n = 1.




"""
```

```python
# Q7: Explain uniform Distribution with an example.
""" uniform distribution is a continuous probability distribution that models the probability of a random variable taking on a 
value within a specified range. In other words, the uniform distribution assumes that all values within the specified 
range are equally likely to occur.

The probability density function (PDF) of the uniform distribution is given by:

f(x) = 1/(b-a) for a ≤ x ≤ b


where a and b are the lower and upper bounds of the range.

An example of the uniform distribution is the probability of rolling a fair six-sided die and obtaining any one of the six
 possible outcomes. In this case, the range of possible outcomes is from 1 to 6, and each outcome is equally likely to occur.
   The PDF of the uniform distribution in this case would be f(x) = 1/6 for 1 ≤ x ≤ 6 and 0 otherwise.

Another example of the uniform distribution is the probability of selecting a random number between 0 and 1.
 In this case, the range of possible outcomes is from 0 to 1, and each value within this range is equally likely to occur.
   The PDF of the uniform distribution in this case would be f(x) = 1 for 0 ≤ x ≤ 1 and 0 otherwise.

"""
```

```python
# Q8: What is the z score? State the importance of the z score.
"""A z-score is a measure of how many standard deviations an observation or data point is away from the mean of its distribution.
 In other words, it tells us how far a value is from the mean in terms of standard deviations.

The formula for calculating the z-score of a value x in a distribution with mean μ and standard deviation σ is:

z = (x - μ) / σ

The importance of the z-score lies in its ability to standardize values from different distributions, making it
 possible to compare them directly. By converting values to z-scores, we can compare values from different distributions
   and determine how extreme or unusual they are relative to the rest of the data. Z-scores can be used to identify outliers 
   or to compare observations from different datasets that have different scales and units of measurement.

 z-scores is used in hypothesis testing, where they can be used to calculate p-values and determine 
the significance of results. They are also used in statistical process control to monitor and analyze process 
performance, where they are used to identify deviations from expected values or trends over time.





"""
```
